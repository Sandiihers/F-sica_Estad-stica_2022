```python
#Creamos una función para obtener un arreglo con número aleatorios entre 0 y 1
# Función para generar números aleatorios
def simulacion_de_volados(n):
    resultados = [0] * n
    for i in range(n):
        resultados[i] = random.randint(0, 1)
    return resultados
```

```python
#Aplicamos la función "simulacion_de_volados" y pedimos que nos realice 100 volados
Simulacion_100=simulacion_de_volados(100)
```

```python
#Definimos t 
t= np.linspace(1,100,100)
```

Obtenemos la frecuencia 

```python
#Definimos una función para obtener la frecuencia
soles = 0
#Arreglo donde se guardará la frecuencia
frecuencia = []
for i in range(len(t)):
    soles = (soles + Simulacion_100[i]) #usamos Simulacion_1000 que es el arreglo donde trae los resultados del volado
    frecuencia.append(soles/t[i])
```

```python
#total de soles
soles
```

```python
frecuencia[len(t)-1]
```

```python
#Como frecuencia[len(t)-1] vemos que la media ideal tiende a 0.5 por lo que haremoss una línea que pase por ahí para
#comparar
media = np.linspace(0.5,0.5,100)
plt.rcParams["figure.figsize"] = (11, 5)
plt.plot(t, frecuencia , color = "green", linewidth = 1.5,label ="Frecuencia Relativa")
plt.plot(t,media, color = "blue", linewidth = 1.5,label ="Media ideal")
plt.title("Frecuencia Relativa para 100 volados")
plt.xlabel("Número de Lanzamiento")
plt.ylabel("Frecuencia Relativa")
plt.legend(loc=0)
plt.show()
```

### Ejercicio 4: Más de caminantes aleatorios... más alcohol

```python
#Generamos los datos aleatorios
x_0 = np.array([-1.,0.,1.])
Caminantes = np.random.choice(x_0,30) 
#ciclo for como Python cuenta el cero entonces es 1000-1=999
for i in range(999): 
    Caminantes = np.vstack((Caminantes, np.random.choice(x_0, 30)))
```

```python
#Arreglo para las posiciones
Posicion = np.empty((1000,30))
for i in range (1000):
    for j in range(30):
        Posicion[i][j] = np.sum(Caminantes[i][:j+1])
```

```python
#Gráficamos
fig1, a_x_1 = plt.subplots(figsize = (11,5))
A_E_X = np.array(range(30))
for i in range(20):
    a_x_1.plot(A_E_X, Posicion[i])
a_x_1.plot(A_E_X, np.sqrt(2/3*A_E_X), c = "black", label = "STD")
a_x_1.plot(A_E_X, -np.sqrt(2/3*A_E_X), c = "black")
a_x_1.set_xlabel("Número de paso")
a_x_1.set_ylabel("Posición")
a_x_1.legend(loc="upper left", fontsize = 15)
plt.show()
```

c) Ahora, considerando los siguientes grupos de listas, calcula el promedio de la posición como función del número de pasos y grafícalo.

i) Grupo 1: Incluye 5 caminatas.

ii) Grupo 2: Incluye 100 caminatas.

iii) Grupo 3: Incluye 500 caminatas.

iv) Grupo 4: Incluye 1000 caminatas.

Por ejemplo, para obtener el promedio de la posición dado el primer paso del Grupo 1, necesitas sumar el primer elemento de cada una de las cincocami natas y dividir la suma entre 5. Para el segundo paso, necesitas sumar los segundos elementos de cada caminata y dividir la suma entre 5; y así sucesivamente (ver Tabla1).

```python
#Caminatas
#Grupo 1
caminatas_grupo1 = 5
#arreglo donde se guardarán los datos
Posicion_promedio1 = np.empty(30)
for j in range(30):
    S = 0
    for i in range(caminatas_grupo1):
        S = S + Posicion[i][j]
        Posicion_promedio1[j] = S / float(caminatas_grupo1)
        
# Grupo 2
caminatas_grupo2 = 100
Posicion_promedio2 = np.empty(30)
for j in range(30):
    S = 0
    for i in range(caminatas_grupo2):
        S = S + Posicion[i][j]
        Posicion_promedio2[j] = S / float(caminatas_grupo2)
        
# Grupo 3
caminatas_grupo3 = 500
Posicion_promedio3 = np.empty(30)
for j in range(30):
    S = 0
    for i in range(caminatas_grupo3):
        S = S + Posicion[i][j]
        Posicion_promedio3[j] = S / float(caminatas_grupo3)

# Grupo 4
caminatas_grupo4 = 1000
Posicion_promedio4 = np.empty(30)
for j in range(30):
    S = 0
    for i in range(caminatas_grupo4):
        S = S + Posicion[i][j]
        Posicion_promedio4[j] = S / float(caminatas_grupo4)
```

```python
A_x_0 = np.linspace(0,30,30)
```

```python
plt.rcParams["figure.figsize"] = (11, 5)
plt.plot(A_x_0,Posicion_promedio1, linewidth = 1.5,label = "5 Caminantes")
plt.plot(A_x_0,Posicion_promedio2, linewidth = 1.5,label = "100 Caminantes")
plt.plot(A_x_0,Posicion_promedio3, linewidth = 1.5,label = "500 Caminantes")
plt.plot(A_x_0,Posicion_promedio4, linewidth = 1.5,label = "1000 Caminantes")
plt.plot(A_x_0,0*Posicion_promedio4, color = "black", linewidth = 1.5, label="Cero")
plt.xlabel("Número de paso")
plt.ylabel("Posición Promedio")
plt.legend(loc=0)
plt.show()
```

```python
# Grupo 1 
caminatas_grupo1_0 = 5
Arreglo1 = np.empty(30)
for j in range(30):
    S = 0
    for i in range(caminatas_grupo1_0):
        S = S + pow(Posicion[i][j]-Posicion_promedio1[j], 2)
        Arreglo1[j] = S / float(caminatas_grupo1_0)

# Grupo 2 
caminatas_grupo2_0 = 100
Arreglo2 = np.empty(30)
for j in range(30):
    S = 0
    for i in range(caminatas_grupo2_0):
        S = S + pow(Posicion[i][j]-Posicion_promedio2[j], 2)
        Arreglo2[j] = S / float(caminatas_grupo2_0)
        
# Grupo 3 
caminatas_grupo3_0 = 500
Arreglo3 = np.empty(30)
for j in range(30):
    S = 0
    for i in range(caminatas_grupo3_0):
        S = S + pow(Posicion[i][j]-Posicion_promedio3[j], 2)
        Arreglo3[j] = S / float(caminatas_grupo3_0)
        
# Grupo 4 
caminatas_grupo4_0 = 1000
Arreglo4 = np.empty(30)
for j in range(30):
    S = 0
    for i in range(caminatas_grupo4_0):
        S = S + pow(Posicion[i][j]-Posicion_promedio4[j], 2)
        Arreglo4[j] = S / float(caminatas_grupo4_0)
```
